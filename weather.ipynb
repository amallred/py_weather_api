```python
def get_astro_info(city, key):
    url = f'{baseurl}astronomy.json?key={key}&q={city}&dt={today}'
    res = requests.get(url)

    if res.status_code == 200:
        data = res.json()
        location = data['location']
        astronomy_data = {
            'city': location['name'],
            'state': location['region'],
            'today' : today,
            'sunrise' : data['astronomy']['astro']['sunrise'],
            'sunset' : data['astronomy']['astro']['sunset'],
            'moonrise' : data['astronomy']['astro']['moonrise'],
            'moonset' : data['astronomy']['astro']['moonset']
        }
        return astronomy_data
    else:
        print(f'Error fetching data for {city}')
```

# Pulling it all together

### `get_current_weather`

```python
def get_current_weather(city, key):
    url = f'http://api.weatherapi.com/v1/current.json?key={key}&q={city}&aqi=yes'
    res = requests.get(url)
    if res.status_code == 200 :
        current_weather_data = res.json()

        location = current_weather_data['location']
        current = current_weather_data['current']
        
        current_weather = {
            'date' : location['localtime'], 
            'location' : location['name'],
            'region' : location['region'],
            'temp_f' : current['temp_f'],
            'temp_c' : current['temp_c'],
            'wind_mph' : current['wind_mph'],
            'humidity' : current['humidity'],
            'air_quality': current['air_quality']['us-epa-index'] #1 = Good, 6 = Hazardous
        }

        return current_weather
    else:
        print(f'Error fetching current weather data for {city}: {res.status_code}')
```

```python
get_current_weather('Amarillo', key)
```

### `get_3_day_forecast`

```python
def get_3_day_forecast (city, key):
    url = f'{baseurl}forecast.json?key={key}&q={city}&days=3&aqi=yes'
    res = requests.get(url)
    if res.status_code == 200 :
        res = requests.get(url)
        data = res.json()
        
        forecast_data = data['forecast']['forecastday']
        
        forecast_3_days = {
            'day': [],
            'location' : [],
            'max_temp_f' : [],
            'max_temp_c' : [],
            'max_wind_mph' : [],
            'humidity' : [],
            'air_quality': []
        }
        for day in forecast_data :

            forecast_3_days['day'].append(day['date'])
            forecast_3_days['location'].append(city)
            forecast_3_days['max_temp_f'].append(day['day']['maxtemp_f'])
            forecast_3_days['max_temp_c'].append(day['day']['maxtemp_c'])
            forecast_3_days['max_wind_mph'].append(day['day']['maxwind_mph'])
            forecast_3_days['humidity'].append(day['day']['avghumidity'])
            forecast_3_days['air_quality'].append(day['day']['air_quality']['us-epa-index'])
        return forecast_3_days
    else:
        print(f'Error fetching 3 day forecast for {city}: {res.status_code}')
```

```python
get_3_day_forecast('Amarillo', key)
```
